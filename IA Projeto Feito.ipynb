###Regressão linear
A regressão linear é uma técnica estatística essencial para modelar a relação entre uma variável dependente e uma ou mais variáveis independentes. Este método assume que a relação entre as variáveis pode ser descrita por uma linha reta, chamada de equação linear. A principal finalidade da regressão linear é prever valores futuros e compreender a estrutura da relação entre as variáveis.

Para realizar a regressão linear, estima-se os coeficientes da equação linear que minimizam a soma dos quadrados das diferenças entre os valores observados e os valores previstos. Este processo é conhecido como método dos mínimos quadrados. A equação resultante pode então ser usada para prever o valor da variável dependente com base em novos valores das variáveis independentes.

Existem duas formas principais de regressão linear: simples e múltipla. A regressão linear simples envolve apenas uma variável independente e uma variável dependente, enquanto a regressão linear múltipla envolve duas ou mais variáveis independentes. Segundo a IBM, a regressão linear é fundamental em áreas como finanças, economia e ciências sociais para análise e previsão de dados, com ferramentas como o IBM SPSS Statistics facilitando sua aplicação. [1]

A TOTVS destaca que a regressão linear é amplamente aplicada em ciência de dados, sendo crucial para transformar dados brutos em insights valiosos. A capacidade de prever e modelar relações entre variáveis torna a regressão linear uma técnica poderosa e versátil. A IBM complementa que a técnica pode lidar com relações tanto lineares quanto não lineares, oferecendo previsões e insights confiáveis, essenciais para a tomada de decisões estratégicas. [2]

###Árvores de Decisão
As árvores de decisão são algoritmos de aprendizado de máquina usados tanto para classificação quanto para regressão. O funcionamento básico envolve a divisão dos dados em subconjuntos menores com base em perguntas que se desdobram em uma estrutura semelhante a uma árvore. Cada nó representa uma pergunta sobre uma característica dos dados e cada folha representa um resultado ou uma classe. As árvores de decisão são conhecidas por serem fáceis de interpretar e visualizar, tornando-as uma ferramenta popular para análises exploratórias de dados. No entanto, elas podem ser propensas ao overfitting, especialmente quando são muito complexas. [3]


###Florestas Aleatórias
A floresta aleatória é um algoritmo de aprendizado de máquina desenvolvido para melhorar a precisão e a robustez das árvores de decisão. Introduzido por Leo Breiman e Adele Cutler, esse método constrói múltiplas árvores de decisão, cada uma treinada com diferentes subconjuntos de dados e características. A principal vantagem dessa abordagem é a redução do overfitting, pois a combinação de várias árvores de decisão diminui a variância do modelo. Na prática, cada árvore na floresta aleatória toma decisões de forma independente e, em seguida, as previsões são agregadas para formar um único resultado final. Isso aumenta significativamente a precisão e a capacidade preditiva do modelo. A IBM enfatiza que a floresta aleatória é altamente eficaz em diversas áreas, como finanças, saúde e marketing, devido à sua capacidade de lidar com grandes volumes de dados e características variadas. [3]

O objetivo é usar essas variáveis como preditoras para construir um modelo que possa prever as notas dos candidatos com base nas suas características demográficas, socioeconômicas e educacionais.

Por fim, é criado uma função que irá realizar o modelo de regressão linear, tal que a formula caclula a variável que queremos prever utilizando os coeficientes que informamos para a regressão. No final, é aplicado o modelo para cada uma das 4 notas.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

df_novo = df.dropna()

colunas_interesse = [
    'TP_FAIXA_ETARIA', 'TP_ESCOLA', 'TP_ENSINO',
    'TP_DEPENDENCIA_ADM_ESC', 'TP_LOCALIZACAO_ESC',
    'Q001', 'Q002', 'Q003', 'Q004', 'Q005', 'Q006', 'Q007', 'Q008', 'Q009', 'Q010',
    'Q011', 'Q012', 'Q013', 'Q014', 'Q015', 'Q016', 'Q017', 'Q018', 'Q019', 'Q020',
    'Q021', 'Q022', 'Q023', 'Q024', 'Q025'
]

X = df_novo[colunas_interesse]
y = df_novo['NU_NOTA_CN']

X = pd.get_dummies(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

modelo_arvore = DecisionTreeRegressor(random_state=42)
modelo_arvore.fit(X_train, y_train)

modelo_floresta = RandomForestRegressor(n_estimators=100, random_state=42)
modelo_floresta.fit(X_train, y_train)

y_pred_arvore = modelo_arvore.predict(X_test)
y_pred_floresta = modelo_floresta.predict(X_test)

rmse_arvore = np.sqrt(mean_squared_error(y_test, y_pred_arvore))
r2_arvore = r2_score(y_test, y_pred_arvore)
rmse_floresta = np.sqrt(mean_squared_error(y_test, y_pred_floresta))
r2_floresta = r2_score(y_test, y_pred_floresta)

print(f'Árvore de Decisão – RMSE: {rmse_arvore}, R²: {r2_arvore}')
print(f'Floresta Aleatória – RMSE: {rmse_floresta}, R²: {r2_floresta}')
```
